```python
import random

brustTimes = []
for i in range(0, 1000):
    brustTimes.append(random.randint(1, 100))

    
priorities = []
for i in range(0, 1000):
    priorities.append(random.randint(1, 5))
    
    
start=1
end=1000
processes=list(range(start,end+1))

process_dict =[]

for i in range (1000):
    process_dict.append({"pro_no":processes[i],"br":brustTimes[i],"pr":priorities[i]})
```

```python

# basic RR

def findWaitingTime(processes, n,wt, quantum):
    rem_bt = [0] * n


    for i in range(n):
        rem_bt[i] = processes[i]["br"]
    t = 0 

   
    while (1):
        done = True

        
        for i in range(n):

            
            if (rem_bt[i] > 0):
                done = False  

                if (rem_bt[i] > quantum):

                   
                    t += quantum

                   
                    rem_bt[i] -= quantum

                
                else:

                  
                    t = t + rem_bt[i]

                   
                    wt[i] = t - processes[i]["br"]

                   
                    rem_bt[i] = 0

        if (done == True):
            break



def findTurnAroundTime(processes, n, wt, tat):
    
    for i in range(n):
        tat[i] = processes[i]["br"] + wt[i]



def findavgTime(processes, n, quantum):
    wt = [0] * n
    tat = [0] * n

    findWaitingTime(processes, n,
                    wt, quantum)

    findTurnAroundTime(processes, n,
                       wt, tat)

    print("Processes    Burst Time     Waiting",
          "Time    Turn-Around Time")
    global total_wt
    total_wt = 0

    total_tat = 0

    for i in range(n):
        total_wt = total_wt + wt[i]
        total_tat = total_tat + tat[i]
        print(" ", processes[i]["pro_no"], "\t\t", processes[i]["br"],
              "\t\t", wt[i], "\t\t", tat[i])

    print("\nAverage waiting time = %.5f " % (total_wt / n))
    print("Average turn around time = %.5f " % (total_tat / n))



if __name__ == "__main__":


  
    n = 1000

    quantum = 20

    findavgTime(process_dict, n, quantum)


```

```python

# algorithm 2 doubling quantum 


def findWaitingTime(processes, n,wt, quantum):
    rem_bt = [0] * n


    for i in range(n):
        rem_bt[i] = processes[i]["br"]
    t = 0 
    
    for i in range(n):


        if (rem_bt[i] > 0):
                

            if (rem_bt[i] > quantum):


                t += quantum

                
                rem_bt[i] -= quantum


            else:


                t = t + rem_bt[i]


                wt[processes[i]["pro_no"] - 1] = t - processes[i]["br"]


                rem_bt[i] = 0
                
                
    sortlist=[]
    for i in range (n):
        sortlist.append({"pro_no":processes[i]["pro_no"],"br":rem_bt[i]})
        
    sort_processes = sorted(sortlist, key=lambda x: x['br'])
    
    rem_bt_1 = [0] * n
    
    for i in range(n):
        rem_bt_1[i] = sortlist[i]["br"]
    
    
    while (1):
        done = True

        
        for i in range(n):

            
            if (rem_bt_1[i] > 0):
                done = False  

                if (rem_bt_1[i] > quantum*2):

                   
                    t += quantum*2

                   
                    rem_bt_1[i] -= quantum*2

                
                else:

                  
                    t = t + rem_bt_1[i]

                   
                    wt[sortlist[i]["pro_no"] -1] = t - processes[i]["br"]

                   
                    rem_bt_1[i] = 0

        if (done == True):
            break



def findTurnAroundTime(processes, n, wt, tat):
    
    for i in range(n):
        tat[i] = processes[i]["br"] + wt[processes[i]["pro_no"] - 1]



def findavgTime(processes, n, quantum):
    wt = [0] * n
    tat = [0] * n

    findWaitingTime(processes, n,
                    wt, quantum)

    findTurnAroundTime(processes, n,
                       wt, tat)

    print("Processes    Burst Time     Waiting",
          "Time    Turn-Around Time")
    global total_wt
    total_wt = 0

    total_tat = 0

    for i in range(n):
        total_wt = total_wt + wt[i]
        total_tat = total_tat + tat[i]
        print(" ", processes[i]["pro_no"], "\t\t", processes[i]["br"],
              "\t\t", wt[processes[i]["pro_no"] -1], "\t\t", tat[processes[i]["pro_no"] -1])

    print("\nAverage waiting time = %.5f " % (total_wt / n))
    print("Average turn around time = %.5f " % (total_tat / n))



if __name__ == "__main__":


  
    n = 1000

    quantum = 20

    findavgTime(process_dict, n, quantum)


```
